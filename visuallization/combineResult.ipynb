```python
import pandas as pd
import matplotlib.pyplot as plt

file_path1 = "/data2/atran16/ProteinClassification_AnhTuanTran/27022026_train_resultTop1/trainingTracking.csv"   # replace with your actual path
df1 = pd.read_csv(file_path1)

file_path2 = "/data2/atran16/ProteinClassification_AnhTuanTran/27022026_train_v2_resultTop1/trainingTracking.csv"   # replace with your actual path
df2 = pd.read_csv(file_path2)

df2["epoch"] = df2["epoch"] + 60

file_path3 = "/data2/atran16/ProteinClassification_AnhTuanTran/27022026_train_v3_resultTop1/trainingTracking.csv"   # replace with your actual path
df3 = pd.read_csv(file_path3)

df3["epoch"] = df3["epoch"] + 100

df = pd.concat([df1, df2, df3], ignore_index=True)

print(df["epoch"].to_list())
print(len(df))
```

```python
plt.figure(figsize=(20, 10)) 
x = (df['epoch']).tolist()

# Create the line chart
plt.plot(x, df["topk1val_acc"] * 100, marker='o', label='top1')
plt.plot(x, df["topk3val_acc"] * 100, marker='o', label='top3')
plt.plot(x, df["topk5val_acc"] * 100, marker='o', label='top5')
plt.plot(x, df["topk10val_acc"] * 100, marker='o', label='top10')
plt.plot(x, df["topk20val_acc"] * 100, marker='o', label='top20')

# Add labels and title
plt.xlabel("epoch")
plt.ylabel("Value(%)")
plt.title("Val Accuracy with topk")
plt.legend()
plt.grid(True)
ticks = list(x[::5])
print(ticks)
if x[-1] not in ticks:
    ticks.append(x[-1])
plt.show()
```

```python
plt.figure(figsize=(20, 10)) 
x = (df['epoch']).tolist()

# Create the line chart
plt.plot(x, df["train_loss"] * 100, marker='o', label='top1')

# Add labels and title
plt.xlabel("epoch")
plt.ylabel("Value(%)")
plt.title("Train Loss")
plt.legend()
plt.grid(True)
ticks = list(x[::5])
print(ticks)
if x[-1] not in ticks:
    ticks.append(x[-1])
plt.show()
```

```python
plt.figure(figsize=(20, 10)) 
x = (df['epoch']).tolist()

# Create the line chart
plt.plot(x, df["val_loss"] * 100, marker='o', label='top1')

# Add labels and title
plt.xlabel("epoch")
plt.ylabel("Value(%)")
plt.title("Val Loss")
plt.legend()
plt.grid(True)
ticks = list(x[::5])
print(ticks)
if x[-1] not in ticks:
    ticks.append(x[-1])
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "/data2/atran16/ProteinClassification_AnhTuanTran/3D_PDB_Dataset/TrainProteinPNG600/8KIB/602.png"

# Read image (BGR)
img = cv2.imread(img_path)

if img is None:
    print("Failed to load image")
else:
    # Resize to 224x224
    img = cv2.resize(img, (224, 224))

    # Convert to RGB for displaying
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Apply lower threshold: <50 → 0, else keep value
    img_masked = img_rgb.copy()
    #img_masked[img_masked < 80] = 0

    # Show image
    plt.figure(figsize=(6,6))
    plt.imshow(img_masked)
    plt.axis("off")
    plt.title("Pixels <50 set to black")
    plt.show()
```
