```python

Resnet50topkAcc = {1:5, 3:16, 5:16, 10:16, 20: 16}
EfficientNetV2topkAcc = {1:5, 3:11, 5:13, 10:22, 20: 27}
ConvNeXttopkAcc = {1:8, 3:10, 5:14, 10:16, 20: 20}
RegNetY16GFtopkAcc = {1:5, 3:7, 5:12, 10:20, 20: 26}

MaxViTtopkAcc = {1:7, 3:10, 5:11, 10:11, 20: 11}
CoAtNettopkAcc = {1:2, 3:12, 5:16, 10:29, 20: 34}
SwinV2BtopkAcc = {1:8, 3:17, 5:20, 10:27, 20: 33}

models = {
    "ResNet50": Resnet50topkAcc,
    "EfficientNetV2": EfficientNetV2topkAcc,
    "ConvNeXt": ConvNeXttopkAcc,
    "RegNetY16GF": RegNetY16GFtopkAcc,
    "MaxViT": MaxViTtopkAcc,
    "CoAtNet": CoAtNettopkAcc,
    "SwinV2B": SwinV2BtopkAcc
}

topk = [1,3,5,10,20]
x = np.arange(len(topk))
width = 0.11

plt.figure(figsize=(12,6))

for i,(name,data) in enumerate(models.items()):
    values = [data[k] for k in topk]
    plt.bar(x + i*width, values, width, label=name)

plt.xticks(x + width*3, topk)
plt.xlabel("Top-K")
plt.ylabel("Accuracy")
plt.ylim(0,40)
plt.title("Top-K Accuracy Comparison Across Models")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.6)

plt.tight_layout()
plt.show()
```
